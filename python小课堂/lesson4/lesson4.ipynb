### 加上标题, 坐标, 注释, 坐标轴含义

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(-10, 11)]
y = [i * i for i in x]

plt.plot(x, y)

plt.show()
```

### 画出图例

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(0, 20)]
y2 = [i * 2 for i in x]
y3 = [i * 3 for i in x]
y4 = [i * 4 for i in x]

plt.title("Function y = x, 2x, 3x, 4x")
for i in [x, y2, y3, y4]:
    plt.plot(i)
plt.xlabel('x')
plt.ylabel('y=f(x)')

plt.show()
```

### 画出不同颜色

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(0, 20)]
y2 = [i + 3 for i in x]
y3 = [i + 6 for i in x]
y4 = [i + 9 for i in x]

plt.title("Function y = x, 2x, 3x, 4x")
# 你可能需要了解一下 以下不同的几种颜色表示方式
for (lines, color) in [(x, 'g'), (y2, '0.5'), (y3, '#FF00FF'), (y4, (0.1, 0.2, 0.3))]:
    plt.plot(lines, color=color)
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```

### 画出不同的线

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(0, 20)]
y2 = [i + 3 for i in x]
y3 = [i + 6 for i in x]

plt.title("Function y = x, 2x, 3x, 4x")
plt.plot(x)
plt.plot(y2)
plt.plot(y3)
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```

### 画出网格图

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(0, 20)]
y2 = [i * 2 for i in x]
y3 = [i * 3 for i in x]
y4 = [i * 4 for i in x]

plt.title("Function y = x, 2x, 3x, 4x")
for i in [x, y2, y3, y4]:
    plt.plot(i)
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.legend(['x', 'y1', 'y2', 'y3'])
plt.show()
```

### 展示局部

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [i for i in range(0, 21)]
y = [i**3 for i in x]

plt.plot(x, y)
# plt.xlim(15, 20)
plt.ylim(400, 800)
plt.show()
```

### 展示日期

```python
import pandas as pd
import matplotlib.pyplot as plt

y = [i for i in range(30)]
x = y # wrong to be fixed

plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.show()
```

### 在一张图中显示不同坐标轴

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

x = [i for i in range(1, 21)]
y1 = [i**2 for i in x]
y2 = [math.log(i) for i in x]

plt.plot(x, y1, 'g')

plt.plot(x, y2, 'r')
# 注意, 这里的y2函数是不是不对?
plt.show()
```

### 填充颜色

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi)
y1 = np.sin(x)
y2 = np.sin(2*x)

plt.plot(x, y1)
plt.plot(x, y2)

plt.title('f(x)=sin x, sin 2x')
plt.show()
```

### 画出不同的形状

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

xy1 = [0.2, 0.2]
xy2 = [0.2, 0.8]
xy3 = [0.8, 0.2]
xy4 = [0.8, 0.8]

fig, ax = plt.subplots()

circle = mpatches.Circle(xy1, 0.15)
ax.add_patch(circle)

ax.axis('equal')
plt.show()
```
